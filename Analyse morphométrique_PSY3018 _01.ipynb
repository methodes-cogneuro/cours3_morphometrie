# Analyses morphométriques

L'objectif d'une analyse morphométrique est d'extraire des **mesures ou quantités** d'intérêt à partir d'images structurelles acquises avec l'**IRM**
- L'exemple ci-contre s'intéresse plus particulièrement au contraste d'une image pondérée en **T1** puisqu'elle donne un bon contraste entre la matière grise, la matière blanche et le liquide céphalo-rachidien

### Application
Une fois une mesure extraite, il est ensuite possible d'étudier les variations qui existent entre différentes populations. E.g. Population atteinte de la maladie de l'Alzheimer par rapport à un groupe de sujets sains 

### Rappel théorique

|**Analyse morphométrique**|**Mesures/quantité extraite**|**Type d'information**|
|----------------------|-------------------------|------------------------------|
|Volumétrie|Volume du cerveau en entier ou d’une ROI (région d’intérêt)|régional|
|VBM|Densité de matière grise *à chaque voxel* du cerveau en entier ou d’une ROI (région d’intérêt)|locale|
|Analyse de surface|Épaisseur cortical et surface *à chaque voxel* du cerveau en entier ou d’une ROI (région d’intérêt)|locale|

### L'intérêt des visualisations morphométriques

- **Visualiser les structures du cerveau**: 

> Visualiser l'aspect du volumes du cerveau ou de ROIs, de la surface cortical, de l'épaisseur cortical. Ce peut-être très pertinent cliniquement : Exemple: peut-on détecter une tumeur (facilité par un contraste T2), une atrophie anormale des tissus ou des ventricules? 


-  **Contrôle de qualité**: 

> Certaines étapes de prétraitement et de l'extraction des mesures ont-elles vraisemblablement échouées? (ex: recalage, artéfacts, segmentation des tissus, etc.)
Le contrôle de qualité est essentiel à la validité des conclusions scientifiques. Il permet d'éviter que le phénomène observé (e.g. différence entre population) soit le résultat d'une erreur/d'artéfacts introduits à l'acquisition ou lors des étapes de prétraitement.  

-  **Cartes statistiques**: 

> Un fois le contrôle de qualité complété, nous effectuons des analyses statistiques sur les mesures extraites à l'échelle du sujet ou du groupe. Dans le cas des analyses statistiques de groupe, nous pouvons être intéressé à représenter les résultats sous la forme d'une carte statistique du cerveau. Ceci est pertinent pour l'analyse en **VBM** et **l'analyse de surface**. 

> *Pourquoi?*  Pour ces deux analyses, nous effectuons un test statistique à CHAQUE voxel d'une région d'intérêt. (= massivement univarié)
Une carte statistique rend cette information plus compréhensible : elle permet d'interpréter chaque résultat statistique relativement à une topographie du cerveau plus accessible (lobe préfrontal, cortex visuel, cortex orbitofrontal, etc.), plutôt que relativement à une coordonnée de l'espace (x, y, z).

###### Prétraitement des cartes de matière grise : lissage

```python
# Initialisation de l'objet nifti_masker
nifti_masker = NiftiMasker(
    standardize = False,
    smoothing_fwhm = 2, # lissage spatiale de l'ordre de 2mm
    memory = 'nilearn_cache')  

# Applique l'objet aux cartes de matière grise 
gm_maps_masked = nifti_masker.fit_transform (gray_matter_map)

hide_toggle()
```

```python
n_samples, n_features = gm_maps_masked.shape
print("%d observations, %d features" % (n_samples, n_features)) 

## Un "feature" correspond à la valeur d'intensité d'un voxel dans l'image. Pour chacun des 50 sujets, nous avons
## 902629 valeurs d'intensité dans la carte de matière grise
hide_toggle()
```

##### Analyses: Pipeline SVR + ANOVA

```python
#### Pipeline prédiction ANOVA et SVR ####

from sklearn.svm import SVR
from sklearn.feature_selection import VarianceThreshold, SelectKBest, \
        f_regression # réduction de la dimensionalité

# Sélection du modèle
svr = SVR(kernel='linear')

#Anova
feature_selection = SelectKBest(f_regression, k=2000)

# Retire les features avec une faible variance inter-sujet 
variance_threshold = VarianceThreshold(threshold=.01)

# Nous avons notre modèle (SVR), notre sélection de features(SelectKBest), 
## et maintenant nous les intégrons dans un pipeline 
from sklearn.pipeline import Pipeline
anova_svr = Pipeline([
            ('variance_threshold', variance_threshold),
            ('anova', feature_selection),
            ('svr', svr)])

### Ajuster le modèle aux données et prédire
anova_svr.fit(gm_maps_masked, age)
age_pred = anova_svr.predict(gm_maps_masked)

hide_toggle()
```

```python
coef = svr.coef_ #Output coefficients 

coef = feature_selection.inverse_transform(coef)
coef = variance_threshold.inverse_transform(coef)
weight_img = nifti_masker.inverse_transform(coef)

hide_toggle()
```

##### Visualisations cartes statistiques

```python
#### Résultats : Poids SVR ####
bg_filename = gray_matter_map[0]
z_slice = 0

fig = plt.figure(figsize=(5.5, 7.5), facecolor='k')

display = plot_stat_map(weight_img, bg_img=bg_filename,
                        display_mode='z', cut_coords=[z_slice],
                        figure=fig, vmax=1)
display.title('Poids SVR', y=1.2) ## changer le titre du graphique
```

Cette figure permet de visualiser les voxels où les variations de la densité de matière grises sont les plus associées à l'âge.  

Au final, nous voulons garder que les voxels pour lesquels la relation est significative (selon un seuil statistique déterminé) C'est ce que nous permet la carte statistique des p-values. 

#### Vérifier l'interprétation @Pierre Bellec 

```python
from nilearn.mass_univariate import permuted_ols

data = variance_threshold.fit_transform(gm_maps_masked)

neg_log_pvals, t_scores_original_data, _ = permuted_ols(
    age, data, 
    n_perm=2000,  
    n_jobs=1)  

signed_neg_log_pvals = neg_log_pvals * np.sign(t_scores_original_data)
signed_neg_log_pvals_unmasked = nifti_masker.inverse_transform(
    variance_threshold.inverse_transform(signed_neg_log_pvals))

hide_toggle()
```

```python
#### Résultats : carte statistique pvalues ####
threshold = -np.log10(0.1)  # 10% corrected

fig = plt.figure(figsize=(5.5, 7.5), facecolor='k')

display = plot_stat_map(signed_neg_log_pvals_unmasked, bg_img=bg_filename,
                        threshold=threshold, cmap=plt.cm.RdBu_r,
                        display_mode='z', cut_coords=[z_slice],
                        figure=fig)
title = ('Negative $\log_{10}$ p-values')
display.title(title, y=1.2)

show()
```

##### Interprétation: 
L'analyse en VBM montre une différence significative de la densité en matière grise en lien avec l'âge pour les voxels...

Limite à cette analyse: On ne contrôle pas certaines covariables importantes comme le score MMSE (indicatif de démence), le genre, etc. Aussi petite taille de l'échantillon.

```python
from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)

hide_toggle()
```
